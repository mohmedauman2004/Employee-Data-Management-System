```python
import csv

class Employee:
    def __init__(self, filename=None, emp_id=None, name=None, position=None, salary=None, email=None):
            self.filename = filename
            self.employees = []
            self.emp_id = emp_id
            self.name = name
            self.position = position
            self.salary = salary
            self.email = email

    def save_data(self):
        with open(self.filename, mode='w') as file:
            writer = csv.writer(file)
            for employee in self.employees:
                writer.writerow([employee.emp_id, employee.name, employee.position, employee.salary, employee.email])

    def valid_salary(self, salary):
        return salary.isdigit()

    def add_employee(self):
        try:
          emp_id = int(input("Enter employee ID "))
        except ValueError:
         print("Incorrect ID")

        for employee in self.employees:
            if employee.emp_id == emp_id:
                print("founded")
                return
        name = input("Enter employee name ")
        position = input("Enter employee position ")

        salary = input("Enter the salary")
        while not self.valid_salary(salary):
            print("incorrect salary")
            salary = input("Enter the salary ")

        email = input("Enter an email ")
        employee = Employee(emp_id, name, position, salary, email)
        self.employees.append(employee)
        self.save_data()
        print("ADDED")

    def show_employees(self):
        for employee in self.employees:
            print(employee)

    def update_employee(self):
      try:
        emp_id = int(input("Enter employee ID "))
      except  ValueError:
       print("Incorrect ID")
       return
      for employee in self.employees:
            if employee.emp_id == emp_id:
                name = input("Enter a new name") or employee.name
                position = input("Enter a new position") or employee.position

                salary = input("Enter the salary")or employee.salary
                while salary and not self.validate_salary(salary):
                    print("incorrect salary")
                    salary = input("Enter the salary") or employee.salary

                email = input("Enter an email") or employee.email
                employee.name = name
                employee.position = position
                employee.salary = salary
                employee.email = email
                self.save_data()
                print("DONE")
                return
      print("Not founded")

    def delete_employee(self):
      try:
        emp_id = int(input("Enter an ID"))
      except ValueError:
        print("Incorrect ID")
        return
        for employee in self.employees:
            if employee.emp_id == emp_id:
                self.employees.remove(employee)
                self.save_data()
                print("DELETED")
                return
        print("NOT FOUNDED")

    def search_employee(self):
      try:
        emp_id = int(input("Enter an ID"))
      except ValueError:
        print("Incorrect ID")
        return
        for employee in self.employees:
            if employee.emp_id == emp_id:
                print("FOUNDED")
                return
        print("NOT FOUNDED")



```
